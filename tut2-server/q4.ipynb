```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
av= +16 ; a_s= 20 ; ac= 0.751 ; asym= 21.4 
mn =  939.57 ; mp =  938.38
c = 3e8
```

```python
def beta_stab(A):
    N = .5*((ac*A**(2/3) - (mn-mp))/(4*asym/A + ac/A**(1/3)) + A)
    Z = .5*(-(ac*A**(2/3) - (mn-mp))/(4*asym/A + ac/A**(1/3)) + A)
    return N, Z
```

```python
def proton_drip(A):
    beta = 2*ac*A**(2/3)/(ac*A**(2/3)+12*asym)
    gamma = (av +3*asym - (2/3)*a_s*A**(-1/3))/((1/3)*ac*A**(2/3) + 4*asym)
    Z = A*(1+beta - np.sqrt((1+beta)**2-gamma))
    N = A- Z 
    return N, Z
```

```python
def neutron_drip(A):
    alpha = (av +3*asym - (2/3)*a_s*A**(-1/3) + (ac/3)*A**(2/3))/((1/3)*ac*A**(2/3) + 4*asym)
    N = A*(1 - np.sqrt(1-alpha))
    return N, A-N
```

```python
A = np.arange(1,300) 
Nb, Zb = beta_stab(A)
Np, Zp = proton_drip(A)
Nn, Zn = neutron_drip(A)
plt.plot(Nb, Zb)
plt.plot(Np, Zp)
plt.plot(Nn, Zn)
plt.show()
```
