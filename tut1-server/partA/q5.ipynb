```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
V0 = 25 # MeV 
a = 0.5 # fm 
R = 5 # fm 
r = np.linspace(-10,10,500)
```

```python
def woods_saxon(r): 
    return V0/(1+np.exp((r-R)/a))
```

```python
plt.plot(r, woods_saxon(r))
plt.xlabel("Distance - r (fm)")
plt.ylabel("Potenial V(r) (MeV)")
```
